```python
import matplotlib.pyplot as plt
u= [4916,4713,3890,3705,3653,3469]
p = [108,44,52.7,50,74,48.8]
cF = [186,66,160,111,143,156]

plt.plot(u,p,color="black",linewidth=3,label="Precio (M)")
plt.plot(u,cF,color="blue",linewidth=3,label="Caballos de Fuerza")
plt.legend()
plt.title("Comparacion Autos vendidos respecto a \n  el Precio/Caballos de Fuerza")
plt.xlabel("Unidades.")
plt.ylabel("Caballos de fuerza/ Precio")
```
